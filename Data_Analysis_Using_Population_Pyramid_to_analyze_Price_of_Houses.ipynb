#example of combining charts

```python
import matplotlib.pyplot as plt

# Sample data for two charts
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values1 = [1000, 2000, 3000, 4000]
values2 = [1500, 2500, 3500, 4500]

# Create figure and axis
fig, ax = plt.subplots()

# Plot the first chart
ax.barh(categories, values1, color='blue', label='Chart 1')

# Plot the second chart with a negative value to display on the left
ax.barh(categories, [-value for value in values2], color='orange', label='Chart 2')

# Add labels, title, and legend
ax.set_xlabel('Values')
ax.set_title('Combined Horizontal Bar Charts')
plt.legend()

# Display the plot
plt.show()
```

#Correlation between price and population (Random Data)

```python
import numpy as np
import scipy.stats as stat

np.random.seed(0)
real_estate_prices = np.random.uniform(100000, 1000000, size=17)  # 17 regions for example


population_data = {
    'Region': ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시',
               '세종특별자치시', '경기도', '강원특별자치도', '충청북도', '충청남도'],
    'Total Population': [686869, 110219, 38833, 29752, 42446, 19915, 20514, 15392, 8746, 206664, 19736, 21503, 29437]
}


real_estate_data = {
    'Region': ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시',
               '세종특별자치시', '경기도', '강원특별자치도', '충청북도', '충청남도'],
    'Real Estate Price': real_estate_prices[:13]  # Only take as many prices as there are regions
}

df_population = pd.DataFrame(population_data)
df_real_estate = pd.DataFrame(real_estate_data)

# Merge the two datasets on the 'Region' column
df_combined = pd.merge(df_population, df_real_estate, on='Region')

# Calculate the correlation between total population and real estate prices
correlation, p_value = stat.pearsonr(df_combined['Total Population'], df_combined['Real Estate Price'])

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 7))
plt.scatter(df_combined['Total Population'], df_combined['Real Estate Price'], color='green')
plt.xlabel('Total Population')
plt.ylabel('Real Estate Price')
plt.title('Population vs Real Estate Price by Region')
plt.grid(True)
plt.show()

# Output the correlation result
correlation, p_value
```

Result : The correlation coefficient is approximately -0.235, indicating a weak inverse relationship between population size and real estate prices for the data we generated. This means that, in our simulated dataset, regions with larger populations tend to have slightly lower real estate prices, but the relationship is not strong
